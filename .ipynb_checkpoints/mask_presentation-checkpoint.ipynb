## Load Mask

The mask comes in the form of a `.tif` file, aka raster format, which we have to convert to a `.nc` file so that we can work with the data using `xarray`.

```python
def load_mask(tif_name):
    nc_file = tif_name + '.nc'
    gdal_ds = gdal.Translate(nc_file,  tif_name + '.tif', format = 'NetCDF')
    xarray_ds = xr.open_dataset(nc_file)
    
    # alert that xr file should be closed xr.close
    print("Please make sure to close the nc file with <xarray_ds>.close()")
    
    return xarray_ds
xarray_final_5bins = load_mask('FINAL_GRID_5BINS')
xarray_final_5bins['Band1'].plot()
```

## Reindex the cutout into mask's resolution and multiply them

```python
def reindex(ds, target):
    return ds.reindex_like(target, method = 'nearest')
```

```python
### define suitable numbers (number or list possible) here

def batch_mask(masks, res_xarray, suitable, mask_col = 'Band1', data_col = 'wind'):
    results = []
    for mask in masks:
        res_xarray = res_xarray.reindex_like(mask, method = 'nearest')
        res_xarray[mask_col] = mask[mask_col]
        if type(suitable) is list:
            res_xarray['mask'] = res_xarray[data_col] * 0
            for i in suitable:
                 res_xarray['mask'] += res_xarray[data_col] * (res_xarray[mask_col] == i)
        elif type(suitable) is int:
            res_xarray['mask'] = res_xarray[data_col] * (res_xarray[mask_col] == suitable)            
        #res_xarray['mask'].plot(cmap = 'bone_r')
        results.append(res_xarray)
    return results
```

```python
def mask(ds_data, ds_mask, col_data, col_mask, col_out, suitable = 1):
    res_ds_data = reindex(ds_data, ds_mask)
    res_ds_data[col_mask] = ds_mask[col_mask]

    if type(suitable) is list:
        res_ds_data[col_out] = res_ds_data[col_data] * 0
        for i in suitable:
            res_ds_data[col_out] += res_ds_data[col_data] * (ds_mask[col_mask] == i)
    elif type(suitable) is int:
            res_ds_data[col_out] = res_ds_data[col_data] * (ds_mask[col_mask] == suitable)
            
    res_ds_data[col_out].plot()
    return res_ds_data
res_xarray_solar = mask(xarray_ds_solar, xarray_final_5bins, 'solar', 'Band1', 'mask', 0)
```

## Get the list of Provinces

```python
def shpreader_path(theme):
    '''
    Given a theme such as 'admin_1_states_provinces' or 'admin_0_countries', download the dbf/shp/shx data from 
    https://www.naturalearthdata.com/downloads/10m-cultural-vectors/;
    
    If the files have been successfully downloaded, return the path to these files that get_shapes() read;
    '''
    return shpreader.natural_earth(resolution='10m', category='cultural',
                                name = theme)

def get_prov_of(country_name):
    '''Take in a country name, return all the shapes of its provinces'''
    reader = shpreader.Reader(shpreader_path('admin_1_states_provinces')) #get provinces path
    provs = []
    prov_names = []
    for r in reader.records():
        if r.attributes['admin'] == country_name:
            provs.append(r.geometry)
            prov_names.append(r.attributes['name'])
    d = {'names': prov_names, 'provinces': provs}
    
    return gpd.GeoDataFrame(d, crs = 'EPSG:4326')

prov_path = shpreader_path('admin_1_states_provinces')
country_path = shpreader_path('admin_0_countries')

provinces = get_prov_of('China')
provinces.head()
```

```python
def grid_coordinates(ds):
    xs, ys = np.meshgrid(ds.coords["lat"], ds.coords["lon"])
    return np.asarray((np.ravel(xs), np.ravel(ys))).T
def grid_cells(ds):
    from shapely.geometry import box
    coords = grid_coordinates(ds)
    span = (coords[len(ds.coords["lon"])+1] - coords[0]) / 2
    hs = np.hstack((coords - span, coords + span))
    out = np.array([])
    for c in hs:
        out = np.append(out, box(*c))
    return 
```

## Seperate the map according to provinces

```python
def shapes_on_cutout(cutout, ds, time, shape, col = 'mask'):
    
    '''
    Input: the cutout object, time (either in string or index),
    and the shape objects we attain from get_shapes()
    Plot a map of the cutout region, with only PM25 data of the specified targets;
    
    Return the indicator matrix.
    
    (The value of the indicator matrix does not effect the PM25 value)
    '''
    
    #set up the axes
    fig = plt.figure(figsize=(10, 5))
    ax = fig.add_subplot(projection=ccrs.PlateCarree());
    ax.set_extent([ds.lon.min(), ds.lon.max(), ds.lat.min(), ds.lat.max()])
    
    #the next 10 lines are for stacking the shape data together, reflecting in the area the cutout covers
    
    #indicator_matrix_ir = np.zeros(ds.shape[1] * ds.shape[2])
    indicator_matrix_ir = np.zeros(len(cutout.indicatormatrix(shape)[0].toarray()))
    
    
    #for i in range(len(shape)):
    indicator_matrix_ir_each = cutout.indicatormatrix(shape)[0].toarray()
        #turn it into a 1D array which we can stack on
        #indicator_matrix_ir = indicator_matrix_ir + indicator_matrix_ir_each.toarray()
        
    indicator_matrix_ir = xr.DataArray(indicator_matrix_ir_each.reshape(cutout.shape), dims=['lat','lon'],
                                       coords=[cutout.coords['lat'], cutout.coords['lon']])
    
    #filter out the non-specified province part
    im = indicator_matrix_ir.copy()
    #im.values = (im.values > 0) * ds.isel(time=time).values()
    im.values = (im.values > 0) * ds.isel(time = 0)[col]
    
        
    ax.coastlines()
    ax.gridlines(draw_labels=True)
    im.plot(ax=ax)
    return im.to_dataset(name = col)

def batch_cutout(cutout, ds, time, shape):
    xarrays_prov = []
    for i in range(len(shape.index)):
        xarrays_prov.append(shapes_on_cutout(cutout, ds, time, [shape.iloc[i]['provinces']]))
    return xarrays_prov
```

## Calculate Areas

```python
def calc_dist(lat1, lon1, lat2, lon2):    
    from math import sin, cos, sqrt, atan2, radians
    # approximate radius of earth in km
    R = 6373.0
    
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c

def calc_area(lat, lat_higher, lon, lon_higher):
    '''
    Turn the area grid into trapezoid and calculate area
    example intput: [35, 35.5, 115, 115.625]
    '''
    
    #generate trapezoid
    top_len = calc_dist(lat_higher, lon, lat_higher, lon_higher)
    bottom_len = calc_dist(lat, lon, lat, lon_higher)
    side = calc_dist(lat, lon, lat_higher, lon)
    
    #calculate the height of the trapezoid
    height = (side ** 2 - ((top_len - bottom_len) / 2) ** 2) ** 0.5
    
    area = (1/2) * (top_len + bottom_len) * height
    return round(area,2)
def calc_prov_mask(xs_provs):
    areas_prov = []
    for xs_prov in xs_provs:
        df = xs_prov.to_dataframe().reset_index(drop = False)
        co_prov = df[df['mask'] > 0][['lat','lon']].values
        area = 0
        for i in co_prov:
            total_area = calc_area(i[0], i[0]+0.5, i[1], i[1]+0.625)
        total_areas.append(total_area)
    return total_areas
```

## Plot the data

```python
def make_figures(xr_prov, col = 'mask'):
    df = xr_prov.to_dataframe().reset_index(drop = False)
    df = df[df[col] > 0]
    co_prov = df[['lat','lon']].values
    areas = []
    for i in co_prov:
        areas.append(calc_area(i[0], i[0]+0.5, i[1], i[1]+0.625))
    df['areas'] = areas
    df['cap'] = df.areas * 27.7 / 1000
    df = df.sort_values(by = col, ascending = False)
    df['cumsum'] = df['cap'].cumsum()
    return df

def batch_figures(xr_provs, provs, col = 'mask'):
    dfs = []
    for xr_prov in xr_provs:
        dfs.append(make_figures(xr_prov, col))
        
    fig, (ax1,ax2, ax3) = plt.subplots(3, 1, figsize=(15, 30))
    for i in range(len(dfs)):
        try:
            if dfs[i][col].iloc[0] > 0.4:
                ax1.plot(dfs[i]['cumsum'], dfs[i][col], label = provs[i])
            else:
                ax2.plot(dfs[i]['cumsum'], dfs[i][col], label = provs[i])
            ax3.plot(dfs[i]['cumsum'], dfs[i][col], label = provs[i])
        except:
            continue
    ax1.set(xlabel='Cumulative Capacity (GW)', ylabel='Capacity Factor (%)',
           title='High Solar Capacity Supply Curves for Provinces in China')
    ax1.legend()

    ax2.set(xlabel='Cumulative Capacity (GW)', ylabel='Capacity Factor (%)',
           title='Low Solar Capacity Supply Curves for Provinces in China')
    ax2.legend()

    ax3.set(xlabel='Cumulative Capacity (GW)', ylabel='Capacity Factor (%)',
           title='Overall Solar Capacity Supply Curves for Provinces in China')
    ax3.legend()
    plt.show()
    return dfs
```

```python
def region_figures(xr_provs, provs, col = 'mask'):
    dfs = []
    regions = provs.grid.unique()
    d = {'E': 'East', 'N':'North', 'C':'Central', 'NW': 'Northwest', 'S': 'South', 'NE':'Northeast', 'XZ':'Xizang'}
    for xr_prov in xr_provs:
        dfs.append(make_figures(xr_prov, col))
        
    fig, axs = plt.subplots(7, 1, figsize=(15, 50))
    for i in range(len(provs)):
        if provs['grid'][i] == regions[0]:
            axs[0].plot(dfs[i]['cumsum'], dfs[i][col], label = provs['names'][i])
        elif provs['grid'][i] == regions[1]:
            axs[1].plot(dfs[i]['cumsum'], dfs[i][col], label = provs['names'][i])
        elif provs['grid'][i] == regions[2]:
            axs[2].plot(dfs[i]['cumsum'], dfs[i][col], label = provs['names'][i])
        elif provs['grid'][i] == regions[3]:
            axs[3].plot(dfs[i]['cumsum'], dfs[i][col], label = provs['names'][i])
        elif provs['grid'][i] == regions[4]:
            axs[4].plot(dfs[i]['cumsum'], dfs[i][col], label = provs['names'][i])
        elif provs['grid'][i] == regions[5]:
            axs[5].plot(dfs[i]['cumsum'], dfs[i][col], label = provs['names'][i])
        elif provs['grid'][i] == regions[6]:
            axs[6].plot(dfs[i]['cumsum'], dfs[i][col], label = provs['names'][i])
    
    for i in range(7):
        axs[i].set(xlabel='Cumulative Capacity (GW)', ylabel='Capacity Factor (%)',
               title='{} Solar Capacity Supply Curves'.format(d[regions[i]]))
        axs[i].legend()
    plt.show()
    return dfs
```
